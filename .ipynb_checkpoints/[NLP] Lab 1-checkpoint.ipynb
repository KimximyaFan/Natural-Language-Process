```python
"""
데이터 셋은 financial_phrasebank를 이용했습니다
8개 중에서 이용하기 제일 편해보였습니다

"""

import pandas as pd
from datasets import load_dataset

dataset = load_dataset("financial_phrasebank", 'sentences_50agree')

# read the dataset
data = dataset['train']

# Convert to pandas DataFrame
df = pd.DataFrame(data)

# sentence 비워져있으면 빈문자열 넣기
df['sentence'].fillna('', inplace=True)

df
```

```python
"""
Stopword 전처리

"""

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stopwords_set = set( stopwords.words("english") )

def filter_stopwords(sentence):
    # Tokenize the sentence
    tokenized_words = word_tokenize(sentence)
    # Filter stopwords, word.lower() 는 소문자 변환
    filtered_words = [word for word in tokenized_words if word.lower() not in stopwords_set]
    # Join the filtered words back into a sentence
    filtered_sentence = ' '.join(filtered_words)
    return filtered_sentence

df['sentence'] = df['sentence'].apply(filter_stopwords)

df
```

```python
"""
label
0 = negative, 
1 = netral, 
2 = positive

데이터 시각화

"""

import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='label', data=df)

plt.xlabel('Sentiment Score')
plt.ylabel('Number of Records')

plt.show()
```

```python
"""
Bag of Words 기법을 사용하여 문장을 특징 벡터로 변환하는 과정

Tokenizer: nltk.tokenize.RegexpTokenizer는 정규식을 사용하여 
문장을 단어로 토큰화하는 역할을 합니다. 
정규식 r'[a-zA-Z]+'은 알파벳으로 이루어진 단어를 추출하는 역할을 합니다. 
이렇게 추출된 단어들은 특징 벡터로 사용합니다

CountVectorizer: sklearn.feature_extraction.text.CountVectorizer는 
문서 집합을 특징 벡터로 변환하는 역할을 합니다. 
각 문서에 대해 단어의 출현 빈도를 세어서 벡터로 표현합니다. 
이때 lowercase=True로 설정하여 모든 단어를 소문자로 변환하고, 
stop_words='english'로 설정하여 영어의 불용어(stopwords)를 제거합니다. 
ngram_range=(1,1)로 설정하여 유니그램(단일 단어)만 고려하고, 
tokenizer=regex_tokenizer.tokenize로 설정하여 앞서 정의한 토크나이저를 사용합니다.

fit_transform: CountVectorizer.fit_transform() 메서드는 문서 집합을 입력으로 받아서 특징 벡터로 변환합니다. 
여기서 df['sentence']는 데이터프레임의 'sentence' 열에 있는 문장들을 포함한 시리즈입니다. 
fit_transform 메서드를 호출하면 문장을 토큰화하고 불용어를 제거하여 단어 출현 빈도를 세고, 
이를 희소 행렬(sparse matrix) 형태로 반환합니다.

최종적으로 count_vectors는 문장들을 단어 출현 빈도에 따라 특징 벡터로 표현한 것을 나타냅니다.

"""

from nltk.tokenize import RegexpTokenizer
from sklearn.feature_extraction.text import CountVectorizer

regex_tokenizer = RegexpTokenizer(r'[a-zA-Z]+')

count_vectorizer = CountVectorizer(lowercase=True,
                     stop_words='english',
                     ngram_range = (1,1),
                     tokenizer = regex_tokenizer.tokenize)

count_vectors = count_vectorizer.fit_transform(df['sentence'])
```

```python
"""
데이터셋 분리

"""

# Partition data into training and testing set
from sklearn.model_selection import train_test_split

feature_train, feature_test, target_train, target_test = train_test_split(
    count_vectors, df['label'], test_size=0.3, random_state=1)
```

```python
"""
모델은 서포트 벡터 머신 이용

SVM은 강력한 분류 모델 중 하나로, 선형 및 비선형 분류 문제에 모두 적용할 수 있습니다. 
SVM은 데이터의 차원이 크고 샘플 수가 적을 때에도 잘 작동하며, 
다중 클래스 분류 문제에도 적용할 수 있습니다.

"""

# import model from scikit-learn model
from sklearn.svm import SVC

# instantiate the model
model = SVC()

# fit the model with data
model.fit(feature_train,target_train)

# Forecast the target variable for given test dataset
predictions = model.predict(feature_test)
```

```python
"""
SVM 모델을 사용하여 분류한 결과의 성능을 측정하는 데 사용되는 여러 가지 지표를 계산합니다. 

Accuracy (정확도):
정확도는 전체 예측 중 올바르게 예측한 비율을 나타냅니다.

Precision (정밀도):
정밀도는 양성 클래스로 예측된 샘플 중 실제 양성 클래스인 비율을 나타냅니다.

Recall (재현율):
재현율은 실제 양성 클래스인 샘플 중에 모델이 정확하게 양성 클래스로 예측한 비율을 나타냅니다.

F1-Score:
F1 점수는 정밀도와 재현율의 조화 평균을 나타냅니다. 이는 정밀도와 재현율의 균형을 고려한 지표입니다.

average='weighted' 매개변수는 다중 클래스 분류 문제에서 
각 클래스의 크기를 고려하여 평균을 계산하도록 지정하는 옵션입니다. 
이는 클래스의 불균형을 고려하여 성능 지표를 계산할 때 유용합니다.

"""

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

# Assess model performance using accuracy measure
print("SVM Model Accuracy:",accuracy_score(target_test, predictions))
# Calculate model precision
print("SVM Model Precision:", precision_score(target_test, predictions, average='weighted'))
# Calculate model recall
print("SVM Model Recall:", recall_score(target_test, predictions, average='weighted'))
# Calculate model f1 score
print("SVM Model F1-Score:", f1_score(target_test, predictions, average='weighted'))
```
